```python
import os

folder_path = r"COCA Dataset\coca-samples-wlp"
output_file = r"COCA Dataset\concatentated_corpus.txt"

def read_and_write_file(input_filename, output_filename, encoding='utf-8'):
    try:
        with open(input_filename, 'r', encoding=encoding) as infile, \
             open(output_filename, 'a', encoding=encoding) as outfile:
            for line in infile:
                outfile.write(line)
    except UnicodeDecodeError:
        print(f"Encoding issue with {input_filename} when using {encoding}")

# Clear the output file first
open(output_file, 'w').close()

# Try reading files with utf-8, then fallback to latin1 if there are errors
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        try:
            read_and_write_file(file_path, output_file, encoding='utf-8')
        except UnicodeDecodeError:
            read_and_write_file(file_path, output_file, encoding='ISO-8859-1')  # Fallback encoding
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Set of English stopwords
stop_words = set(stopwords.words('english'))

def preprocess_text(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    processed_lines = []
    for line in lines:
        # Strip line numbers and other metadata, keeping only the main text content
        parts = line.split('\t')
        if len(parts) > 3:
            content = parts[2]  
            content = content.lower()
            tokens = word_tokenize(re.sub(r'[^\w\s]', '', content))
            filtered_tokens = [word for word in tokens if word not in stop_words]
            processed_lines.append(' '.join(filtered_tokens))

    # Save processed lines to output file
    with open(output_file, 'w', encoding='utf-8') as file:
        for line in processed_lines:
            file.write(line + '\n')


# File paths
input_file_path = r"COCA Dataset\concatentated_corpus.txt"
output_file_path = r"COCA Dataset\big_dataset.txt"

# Preprocess the text
preprocess_text(input_file_path, output_file_path)
```

```python
# Path to your processed text file
file_path = r"COCA Dataset\big_dataset.txt"


# Read data from file
sentences = []
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        words = line.strip().split()
        if words:  # ensuring no empty lines
            sentences.append(words)

# Check the first few entries
print(sentences[:10])
```

```python
from gensim.models import Word2Vec
import logging

# Enable logging to monitor training
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# File path
file_path = r"COCA Dataset\big_dataset.txt"

# Read data
def read_input(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.split()

# This will stream lines from the file one by one
sentences = list(read_input(file_path))

# Train a Word2Vec model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4, sg=1)

# Save the model
model.save("word2vec.model")

print("Model training completed.")
```

```python
# Load the model
model = Word2Vec.load("word2vec.model")
word='smart'
# Get the vector for a word
vector = model.wv[word]  # replace 'example' with any word

# Find similar words
similar_words = model.wv.most_similar(word, topn=10)  # replace 'example'

print(f"Vector for '{word}':", vector)
print(f"Similar words to '{word}':", similar_words)
```

```python
from gensim.models import Word2Vec

# Load the model
model = Word2Vec.load("word2vec.model")

# Function to calculate similarity between two words
def get_similarity(word1, word2):
    # Check if both words are in the vocabulary
    if word1 in model.wv and word2 in model.wv:
        # Calculate similarity
        similarity = model.wv.similarity(word1, word2)
        return similarity
    else:
        return "One or both words not in the vocabulary!"

# Example usage
word1 = 'smart'  # Replace 'example1' with your word
word2 = 'intelligent' # Replace 'example2' with your word

similarity_score = get_similarity(word1, word2)
print(f"Similarity between '{word1}' and '{word2}':", similarity_score)
```
